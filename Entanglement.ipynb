QUANTUM ENTANGLEMENT

A set of qubits whose combined state is |w> is said to be in state of entanglement if |w> cann't be decomposed as |w> = |u> * |v> ,where |u> and |v> are two independent quantum states. ( * represents tensor product )

Creating an Entanglement pair (|00> + |11>)/sqrt(2)

```python
#importing the required classes and modules.
from qiskit import QuantumRegister, ClassicalRegister
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram

#creating quantum register and quantum circuit objects
q = QuantumRegister(2)
c = ClassicalRegister(2)
qc = QuantumCircuit(q,c)

#Applying the entangling operation
qc.h(q[0])  # Applying Hadamard gate on the first qubit
qc.cx(q[0],q[1])  # Applying CNOT gate, where the first qubit is the controlled qubit and second one is target qubit
qc.measure(q,c)

#Executing the circuit in the simulator
backend = Aer.get_backend('qasm_simulator')
qjob = execute(qc, backend, shots = 1000)
count = qjob.result().get_counts()
print(count)

qc.draw()
```

```python
# Histogram ( It will help us to observe the probabilty of qubits in entagled pair)
plot_histogram(count)
```



Creating an Entanglement pair (|000> + |111>)/sqrt(2)

```python
#importing the required classes and modules.
from qiskit import QuantumRegister, ClassicalRegister
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram

#creating quantum register and quantum circuit objects
q = QuantumRegister(3)
c = ClassicalRegister(3)
qc = QuantumCircuit(q,c)

#Applying the entangling operation
qc.h(q[0])  # Applying Hadamard gate on the first qubit
qc.cx(q[0],q[1])  # Applying CNOT gate, where the first qubit is the controlled qubit and second one is target qubit
qc.cx(q[1],q[2])
qc.measure(q,c)

#Executing the circuit in the simulator
backend = Aer.get_backend('qasm_simulator')
qjob = execute(qc, backend, shots = 1000)
count = qjob.result().get_counts()
print(count)

qc.draw()
```

```python
# Histogram ( It will help us to observe the probabilty of qubits in entagled pair)
plot_histogram(count)
```
